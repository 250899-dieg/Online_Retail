# Limpieza de datos


```python
"""Aqui vamos a crear un nuevo data frame donde vamos a guardar los datos de
limpieza"""
data_cleaned = data.drop_duplicates() #Aqui le decimos que elimine los datos duplicados del dataframe
data_cleaned = data_cleaned.dropna(subset=['CustomerID']) #Aqui le digo que me elimine los datos faltantes de la columna tal...
```

```python
#Devolvemos  el número de valores nulos por columna
data_cleaned.isnull().sum()
```

```python
#De todo el set de datos miramos cuantos datos tenemos duplicados
data_cleaned.duplicated().sum()
```

# Ajuste de dataframe

```python
data_cleaned
```

```python
#Aca lo que estamos haciendo es crear una nueva columna, donde le decimos que el valor va a ser la columna cantidad por precio unitario
data_cleaned['TotalAmount'] = data_cleaned['Quantity'] * data_cleaned['UnitPrice']
data_cleaned.head()
```

```python
#Vamos a hacer una conversion de serie de tiempo, cambiamos el tipo de dato
data_cleaned['InvoiceDate'] = pd.to_datetime(data_cleaned['InvoiceDate'])
data_cleaned.head()
```

```python
# Desglosar la columna 'InvoiceDate' en columnas temporales
data_cleaned['Year'] = data_cleaned['InvoiceDate'].dt.year
data_cleaned['Month'] = data_cleaned['InvoiceDate'].dt.month

# Mostrar las primeras filas con las nuevas columnas temporales
data_cleaned.head()
```

# Analisis de dataframe


```python
#Aca hacemos un grupo de ventas por año, le decimos que nos de la suma de totalamount por año
sales_by_year = data_cleaned.groupby('Year')['TotalAmount'].sum()
sales_by_year
```

```python
data_cleaned['Semester'] = data_cleaned['Month'].apply(lambda x:1 if x<=6 else 2)
sales_by_semester = data_cleaned.groupby(['Year', 'Semester'])['TotalAmount'].sum()
sales_by_semester
```

```python
#devoluciones o retornos de productos.
total_returns = data_cleaned[data_cleaned['Quantity'] < 0].shape[0]
total_returns
```

```python
total_non_returns = data_cleaned[data_cleaned['Quantity'] >= 0].shape[0]
total_non_returns
```

# Validacion de graficas


```python
import matplotlib.pyplot as plt

labels = ['Devoluciones', 'No Devoluciones']
sizes = [total_returns, total_non_returns]
colors = ['lightcoral', 'lightgreen']

plt.figure(figsize = (8,8))
plt.pie(sizes, labels=labels, colors= colors, startangle=140)

plt.title('Porcentaje de transacciones con y sin devolución')
plt.show()
```

```python
# Crear una columna categórica basada en el monto total de la transacción (ejemplo: 'Low', 'Medium', 'High')
def categorize_total_amount(amount):
    if amount < 20:
        return 'Low'
    elif 20 <= amount < 100:
        return 'Medium'
    else:
        return 'High'

data_cleaned['AmountCategory'] = data_cleaned['TotalAmount'].apply(categorize_total_amount)

# Mostrar las primeras filas con las nuevas columnas
print(data_cleaned.head())
```

```python
plt.figure(figsize=(12,6))
data_cleaned.groupby(['Year', 'Month'])['TotalAmount'].sum().plot(kind='bar')
plt.title('Distribución de ventas por mes y año')
plt.xlabel('Año, Mes')
plt.ylabel('Ventas totales')
plt.show()
```

```python
top_products = data_cleaned.groupby('StockCode')['Quantity'].sum().sort_values(ascending=False).head(10)
top_products = top_products.reset_index()
top_products = pd.merge(top_products, data_cleaned[['StockCode', 'Description']].drop_duplicates(),
                        on='StockCode', how='left')
```

```python
plt.figure(figsize=(12,8))
plt.barh(top_products['Description'],top_products['Quantity'])
plt.title('Top de productos')
plt.xlabel('Cantidad vendida')
plt.ylabel('Producto')
plt.gca().invert_yaxis()

plt.show()
```
